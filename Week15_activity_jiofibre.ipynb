```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/jio fibre tcp-connection.csv')
data.shape
```

```python
data.head()
```

```python
data.duplicated().sum()
```

```python
data = data.drop_duplicates()
data.shape
```

```python
data.isnull().sum()
```

```python
data.info()
```

```python
for col in data.columns:
  print(data[col].value_counts())
```

```python
data.dropna(how='any', inplace=True)
data.shape
```

```python
def encode(data):
  for col in data.columns:
    data[col] = data[col].factorize()[0]
  return data
```

```python
encoded_data = encode(data.copy())
encoded_data.head()
```

```python
encoded_data['protocol_type'].unique()
```

##splitting into target and feature variables.

```python
x = encoded_data.drop('intrusion', axis=1)
y = encoded_data['intrusion']
```

### splitting into train and test data.

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()

tree.fit(x_train, y_train)
y_pred = tree.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print('\n', accuracy_score(y_test, y_pred))
```
